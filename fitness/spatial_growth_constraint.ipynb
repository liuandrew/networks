# The following code will be used to generate a "spatial growth constraint model" (tentative name)

## Process

* (G.) Denotes possible generalizations
* (P.) Denotes tweaking parameters

**a.** Start with a seed network (random geometric graph) with n nodes positioned random uniform in the space of the network, with each possible edge formed independently with probability p

Next is the growth stage of the network: repeat the following until the desired network size is achieved

**b.** Select a node v in the existing network (randomly, or (G.) with respect to degree or a fitness)

**c.** Place a new node u spatially somewhere in a circle of radius (P.) r around v (randomly select distance d and angle theta for 2D)

**d.** Add edge u-v

**e.** For all nodes w (P.) mu jumps away from u, and within (P.) nu distance of u, add edge u-w

## Alternative  spatial closeness constraints

We want to consider other limitations as well. I think that it would be best to choose one from each of the following categories, mixing and matching.

#### Edge / other limits
* Number of edges that an incoming node can form
* Fitness

#### Spatial limits
These are limits in the space that nodes are embedded in
* Limit distance away
* Fitness times distance modifier

#### Network limits
These are limits in how far along the network the nodes can probe
* Number of jumps away
* Distance along network links away

I guess you don't need a seed node to use network limits, you could just attach the node to the closest/most fit node and then work from there? Could be another interesting model and allows us to use random placement

Combinations:
* Num edges / Distance / Jumps
* Num edges / Distance along network links

```python
class LabelledHeap():
  '''
  A labelled heapq, using min heap
  The root most element (index 0) is by definition the smallest element
  Each child element (index 2N+1 and 2N+2, since we are zero indexed) 
    must be greater or equal to the parent

  items: dict with keys representing labels and values as positions in the heap
  heap: the heap
  '''
  def __init__(self):
    self.items = {}
    self.heap = []

  def push(self, value, label):
    '''
    Find the correct place for a new entry based on its attributed value
    If any swaps are made, update each dictionary position as it is passed
    '''
    if(label in self.items):
      raise Exception('That item already exists')

    new_entry = (value, label)
    self.heap.append(new_entry)
    pos = len(self.heap) - 1
    self.items[label] = pos

    self._siftdown(0, pos)

    
  def update(self, value, label):
    '''
    Update a node with the given value, then check to sift it into the correct
    positions
    '''
    pos = self.items[label]
    item = (value, label)
    self.heap[pos] = item

    self._siftup(pos)

  def pop(self):
    '''
    Remove the smallest item in the heap, updating dictionary positions while
    doing so
    '''
    lastelt = self.heap.pop()
    if(self.heap):
      returnitem = self.heap[0]
      self.heap[0] = lastelt
      self._siftup(0)
    else:
      returnitem = lastelt
    
    del self.items[returnitem[1]]

    return returnitem


  def _siftdown(self, startpos, pos):
    '''
    Given the starting position of a element, move the element into
    the correct place, making comparisons with parent elements
    '''
    item = self.heap[pos]
    value = item[0]
    label = item[1]
    
    while(pos > 0):
      parentpos = (pos - 1) >> 1
      parent = self.heap[parentpos]
      parent_value = parent[0]
      parent_label = parent[1]
      #check to see if item is greater and should continue to sift
      if(self.cmp_lt(value, parent_value)):
        self.heap[pos] = parent
        self.items[parent_label] = pos
        pos = parentpos
        continue
      break

    self.heap[pos] = item
    self.items[label] = pos


  def _siftup(self, pos):
    '''
    Given the starting position of an element, move the element into the correct
    place, making comparisons with child elements
    '''
    endpos = len(self.heap)
    startpos = pos
    newitem = self.heap[pos]
    # Bubble up the smalelr child until hitting a leaf
    childpos = 2*pos + 1
    while(childpos < endpos):
      rightpos = childpos + 1
      if(rightpos < endpos 
          and not self.cmp_lt(self.heap[childpos][0], self.heap[rightpos][0])):
        childpos = rightpos
      # Move the smaller child up
      self.heap[pos] = self.heap[childpos]
      child_label = self.heap[childpos][1]
      self.items[child_label] = pos
      pos = childpos
      childpos = 2*pos + 1

    # The leaf at pos is empty now. Put newitem there and bubble it up to its
    # final resting place (by sifting parents down)
    self.heap[pos] = newitem
    self._siftdown(startpos, pos)



  def cmp_lt(self, x, y):
    # Use __lt__ if available; otherwise, try __le__.
    # In Py3.x, only __lt__ will be called.
    return (x < y) if hasattr(x, '__lt__') else (not y <= x)
  
```

```python
def coordinate_dijkstras(G, source, max_distance=False, return_sorted=False):
    '''
    Get the shortest path length from a source to all other nodes using Dijkstra's algorithm
    Assumes that the graph is a SimpleSpatialGraph i.e. nodes are labelled using incrementing
        numbers and each node has a 'coordinate' key
    All nodes with infinite distance will be removed from the shortest_paths dictionary
    
    Optionally pass a max_distance to cut off search once a distance is exceeded
    Optionally pass sorted=True to return a list of tuples of nodes sorted by distance
        NOTE: The return is no longer a dictionary but a sorted list of tuples
    '''
    shortest_paths = dict()
    visited = list()
    q = LabelledHeap()
    for i in range(G.node_count):
        shortest_paths[i] = 9999999999
        visited.append(False)
    shortest_paths[source] = 0
    q.push(0, source)
    
    while(len(q.heap)):
        current_distance, node = q.pop()
        
        if(max_distance and current_distance > max_distance):
            break
        
        if(not visited[node]):
            visited[node] = True
            coordinate = g.nodes[node]['coordinate']
            adjacents = list(g[node])
            
            for adj_node in adjacents:
                if(not visited[adj_node]):
                    adj_coordinate = g.nodes[adj_node]['coordinate']
                    dist = distance.euclidean(coordinate, adj_coordinate)
                    total_dist = current_distance + dist
                    
                    #If the new path is shorter than the most 
                    if(shortest_paths[adj_node] > total_dist):
                        shortest_paths[adj_node] = total_dist
                        if(adj_node in q.items):
                            q.update(total_dist, adj_node)
                        else:
                            q.push(total_dist, adj_node)
            
    #remove nodes with infinite distance
    shortest_paths = {k:v for k, v in shortest_paths.items() if v != 9999999999}
    
    #remove nodes with distance greater than max_distance
    if(max_distance):
        shortest_paths = {k:v for k, v in shortest_paths.items() if v <= max_distance}
    
    if(return_sorted):
        shortest_paths = sorted(shortest_paths.items(), key=lambda x: x[1])
        
    return shortest_paths
```
