```python
'''
compute partition reverse optimizations upto 10 nodes from every starting node
'''

load = '50_uniform_1'

if(load):
    res = pickle.load(open(load + '/res', 'rb'))
#     res = pickle.load(open(load + '/res2', 'rb'))
    
reverse_optimized_parts_10 = []
for i in range(50):
    if(i % 10 == 0):
        print(i)
    optimization_res = reverse_greedy_optimization(res, i, 10, interval=(1000, None))
    reverse_optimized_parts_10.append(optimization_res[0])
    plt.scatter(np.full(10, i), optimization_res[0])
```

```python
plt.figure(figsize=(15,10))

load = '50_uniform_1'
partitions = ['reverse_optimized_parts_10', 'reverse_optimized_parts_10_interval',
             'reverse_optimized_parts_10_2', 'reverse_optimized_parts_10_interval_2']

baseline = pickle.load(open(load + '/' + partitions[0], 'rb'))

for i in range(50):
    plt.scatter(np.full(10, i * 4), baseline[i], color='blue')

for p, part in enumerate(partitions):
    if(p == 0):
        continue
    partition = pickle.load(open(load + '/' + part, 'rb'))
    for i in range(50):
        matched = []
        extra = []
        
        for j in range(10):
            if(partition[i][j] in baseline[i]):
                matched.append(partition[i][j])
            else:
                extra.append(partition[i][j])
        
        plt.scatter(np.full(len(matched), i * 4 + p), matched, color='green')
        plt.scatter(np.full(len(extra), i * 4 + p), extra, color='red')
```

```python
plt.figure(figsize=(15,10))

load = '50_uniform_1'
partitions = ['reverse_optimized_parts_10', 'reverse_optimized_parts_10_interval',
             'reverse_optimized_parts_10_2', 'reverse_optimized_parts_10_interval_2']

baseline = pickle.load(open(load + '/' + partitions[0], 'rb'))

for p, part in enumerate(partitions):
    if(p == 0):
        continue
    partition = pickle.load(open(load + '/' + part, 'rb'))
    total_matched = []
    
    for i in range(50):
        matched = 0        
        for j in range(10):
            if(partition[i][j] in baseline[i]):
                matched = matched + 1
        total_matched.append(matched)
    plt.scatter(range(50), total_matched, label=p, marker=p + 8)
    print(np.sum(total_matched) / 500)

plt.legend()

        
```

```python
np.argwhere(np.array(total_matched) == 9)
```

```python
'''
symmetricity score of the above plot. symmetricity calculated two numbers
positive: how many partitions it was part of (horizontal) that were not in the vertical
negative: how many nodes it had in partition (vertical) that it was not a part of in horizontal
'''
reverse_optimized_sym_10_x = {}
reverse_optimized_sym_10_y = {}

for i in range(50):
    reverse_optimized_sym_10_x[i] = []
    reverse_optimized_sym_10_y[i] = []
    
for i in range(50):
    for j in reverse_optimized_parts_10[i]:
        reverse_optimized_sym_10_x[j].append(i)
        reverse_optimized_sym_10_y[i].append(j)

sym_diff_x = []
sym_diff_pos = []
sym_diff_neg = []
for i in range(50):
    positive_count = 0
    negative_count = 0
    for j in reverse_optimized_sym_10_x[i]:
        if(j not in reverse_optimized_sym_10_y[i]):
            positive_count = positive_count + 1
    for j in reverse_optimized_sym_10_y[i]:
        if(j not in reverse_optimized_sym_10_x[i]):
            negative_count = negative_count - 1
    sym_diff_pos.append(positive_count)
    sym_diff_neg.append(negative_count)

plt.figure(figsize=(10,10))
plt.scatter(range(50), sym_diff_pos, c='green', marker='D')
plt.scatter(range(50), sym_diff_neg, c='red', marker='x')
```
